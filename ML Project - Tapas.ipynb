# Reading Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading CSV File

```python
df = pd.read_excel('zomato.xlsx')
df.head()
```

```python
df.shape
```

```python
#get the data info of the columns
df.info()
```

```python
#get the list of all the columns
df.columns
```

# Cleaning Data

```python
#check for any duplicate values
df.duplicated().sum()
```

```python
#check for missing values
df.isnull().sum()
```

```python
#Dropping unnecessary columns

df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list','listed_in(city)'], axis = 1)
df.head()
```

```python
# Renaming the columns
df=df.rename(columns={"name":'Name','rate':'Ratings','votes':'Votes','rest_type':'Rest_Type','cuisines':'Cuisines',
                      'approx_cost(for two people)':'Cost','listed_in(type)':'Type','location':'City',
                      'online_order':'Takes online orders?','book_table':'Has table booking?'})
df.head()
```

```python
sum(df.duplicated())
```

```python
# Dropping Duplicates
df.drop_duplicates(inplace = True) 
df.shape
```

```python
#check for missing values for remaining columns
df.isnull().sum()
```

```python
df['Ratings'].unique()
```

```python
# Removing "NEW" , "-" and "/5" from Rate Column
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['Ratings'] = df['Ratings'].apply(handlerate)
df['Ratings'].head()
```

```python
# Filling Null Values in Rate Column with Mean

df['Ratings'].fillna(df['Ratings'].mean(), inplace = True)
df['Ratings'].isnull().sum()
```

```python
df.info()
```

```python
# Dropping Null Values
df.dropna(inplace = True)
df.head()
```

```python
df['Cost'].unique()
```

```python
# Removing , from Cost Column

def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost'] = df['Cost'].apply(handlecomma)
df['Cost'].unique()
```

```python
df.head()
```

```python
df.isnull().sum()
```

# Plot the count of rating.

```python
plt.rcParams['figure.figsize'] = 14,7
sns.countplot(df["Ratings"], palette="Set1")
plt.title("Count plot of rate variable",size=20)
plt.xticks(rotation = 90)
plt.show()
```

## The rate variable follows near normal distribution with mean equal to 3.7. The rating for the majority of the restaurants lies within the range of 3.5-4.2. Very few restaurants (~350) has rated more than 4.8. 

# Analyze the number of Restaurants in Cities

```python
df.City.value_counts().nlargest(10).plot(kind = "barh")
plt.title("Number of restaurants by City")
plt.xlabel("Counts of Cities",size=10)
plt.ylabel('Cities Name',size=10)
plt.show()
```

## Most of the restaurants are located in BTM Layout area, makes it one of the most popular residential and commercial places in Bangalore.

# How many Restaurants in Bangalore take Online Orders?

```python
print(df['Takes online orders?'].value_counts())
plt.figure(figsize=(20,10))
df['Takes online orders?'].value_counts().plot(kind='pie',colors=['lightblue','skyblue'],autopct='%1.1f%%', textprops={'fontsize': 20})
plt.title('% of Restaurants that take Online Orders',size=20)
plt.xlabel('',size=10)
plt.ylabel('',size=10)
plt.legend(loc=2, prop={'size': 15})
```

## The above graph that in nearly 66% of restaurants, an online ordering facility is available.

# Analyzing the Restaurants based on availability of Online Order Facility 

```python
#count plot for online_order analysis
sns.countplot(df["Takes online orders?"], palette = "Set1")
plt.xlabel('',size=10)
plt.ylabel('',size=10)
plt.show()
```

## More than 60% of the Restaurants listed in Zomato provide an option of Online Order remaining Restaurants has an option of Dine-In only.

# Does Online Order facility impacts the Rating of the Restaurant?

```python
sns.countplot(hue = df["Takes online orders?"], palette = "Set2", x = df["Ratings"])
plt.title("Distribution of Restaurant rating over Online Order Facility",size=20)
plt.xticks(rotation = 90)
plt.xlabel('Rating',size=15)
plt.ylabel('Count',size=15)
plt.legend(loc=2, prop={'size': 15})
```

## Restaurants which provides Online Order facitity has better Ratings than the Traditional Restaurants.

# What percentage of Restaurants offer Table Booking Facilities?

```python
print(df['Has table booking?'].value_counts())
plt.figure(figsize=(30,10))
df['Has table booking?'].value_counts().plot(kind='pie',colors=['plum','mediumorchid'],autopct='%1.1f%%', textprops={'fontsize': 15})
plt.title('% of Restaurants that provide Table Booking Facility',size=20)
plt.xlabel('',size=15)
plt.ylabel('',size=15)
plt.legend(loc=2, prop={'size': 15})
```

# What was the most common Rating?

```python
ratings=df.groupby(['Ratings']).size().reset_index().rename(columns={0:"Rating_Count"})
plt.figure(figsize=(30,10))
sns.barplot(x='Ratings',y='Rating_Count',data=ratings)
plt.title('Rating vs Rating counts',size=40)
plt.xlabel('Ratings',size=30)
plt.ylabel('Ratings Count',size=30)
```

## Most of the Restaurants in Bangalore received a rating of 3.6 to 4. Very few Restaurants have Poor Ratings, and quite a group of Restaurants has Excellent Ratings of 4.9 or 5.

# Is there any Correlation between the two peoples approximated cost and a Restaurants Ratings?

```python
plt.figure(figsize=(30,10))
sns.lmplot(x='Ratings',y='Cost',data=df,height=7)
plt.xlabel('Ratings',size=15)
plt.ylabel('Cost for two people',size=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in current_values])
```

## As we can see, restaurants that cost less have better reviews than restaurants that are expensive.

# Which are the Biggest Restaurant Chain and Best Restaurant Chain ?

```python
plt.rcParams['figure.figsize'] = 14,7
plt.subplot(1,2,1)
df.Name.value_counts().head().plot(kind = "barh", color = sns.color_palette("hls", 5))
plt.xlabel("Number of restaurants",size=15)
plt.title("Biggest Restaurant Chain (Top 5)",size=15)
plt.subplot(1,2,2)
df[df['Ratings']>=4.5]['Name'].value_counts().nlargest(5).plot(kind = "barh", color = sns.color_palette("Paired"))
plt.xlabel("Number of restaurants",size=15)
plt.title("Best Restaurant Chain (Top 5) - Rating More than 4.5",size=15)
plt.tight_layout()
```

## Cafe Coffee Day chain has over 90 cafes across the city that are listed in Zomato. On the other hand, Truffles  a burger chain has the best fast food restaurants (rating more than 4.5 out of 100), quality over quantity.
